```python
df['Date'] = pd.DatetimeIndex(df['ActivityDate'])
df['Day'] = pd.DatetimeIndex(df['ActivityDate']).day
df['dayofweek'] = df['Date'].dt.day_name()
df['Month'] = pd.DatetimeIndex(df['ActivityDate']).month
df['Year'] = pd.DatetimeIndex(df['ActivityDate']).year
```

```python
# drop the Activitydate 
df.drop(['ActivityDate'],axis =1)
```

### Seaborn and Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Create box plot

```python
# Box plot using Date
plt.figure(figsize = (30,10))
plt.xticks(rotation=90)
plt.rc('font', size=20)
sns.boxplot(x = 'Date', y = 'Calories', data = df)
```

```python
# Box plot using dayofweek
plt.figure(figsize = (30,10))
plt.rc('font', size=15)
sns.boxplot(x = 'dayofweek', y = 'Calories', data = df, order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
```

#### Scatter Plot

```python
# To find correlation between two features
plt.figure(figsize=(30,10))
plt.rc('font', size=15)
sns.scatterplot(x='Calories',y='TotalSteps',data=df)
```

#### Correlation Matrix (Heatmap)

```python
plt.figure(figsize=(30,30))
plt.rc('font', size=15)
df_corr = sns.heatmap(df.corr(),annot=True)
plt.show()
```

### Distribution plot 

```python
plt.figure(figsize=(5,5))
plt.rc('font', size=7.5)
sns.distplot(df['Calories'])
```
